```python
import json
import urllib.request
import pandas as pd
```

```python
def get_search_url(start_date, end_date, pages, limit):
    base = 'https://open.jejudatahub.net/api/proxy/5D5a577taba7tbb71at1b1bt9tatata9/'
    start = start_date
    end = end_date
    client_key = 'b411t4bjr488rot1bp33t4_428teo32j'
    
    param_date = "?startDate=" + start_date + "&endDate=" + end_date 
    request_param = "&number=" + pages + "&limit=" + limit
    
    return base + client_key + param_date + request_param
```

```python
def get_result(url):
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    
    return json.loads(response.read().decode('utf-8'))
```

```python
def get_fields(json_data):
    sido = [each['sido'] for each in json_data['data']]
    sigungu = [each['sigungu'] for each in json_data['data']]
    cityGubun = [each['cityGubun'] for each in json_data['data']]
    marketType = [each['marketType'] for each in json_data['data']]
    userType = [each['userType'] for each in json_data['data']]
    ageGroup = [each['ageGroup'] for each in json_data['data']]
    gender = [each['gender'] for each in json_data['data']]
    dtYearMonth = [each['dtYearMonth'] for each in json_data['data']]
    userCount = [each['userCount'] for each in json_data['data']]
    useCount = [each['useCount'] for each in json_data['data']]
    useCost = [each['useCost'] for each in json_data['data']]
    
    data_pd = pd.DataFrame({'sido':sido,'sigungu':sigungu,'marketType':marketType,'userType':userType,'gender':gender,'userCount':userCount,'useCount':useCount,'useCost':useCost},
                          columns = ['sido','sigungu','marketType','userType','gender','userCount','useCount','useCost'])
    
    return data_pd
```

```python
def add_col_row():
    pd.set_option('display.max_row', 500)
    pd.set_option('display.max_columns', 100)
    
add_col_row()
```

```python
res_jeju_data = pd.DataFrame()

page_count = 1
limit = 100
start_date = '20170101'
end_date = '20181231'

while True:
    
    url = get_search_url(start_date, end_date, str(page_count), str(limit))
    json_data = get_result(url)
    pd_result = get_fields(json_data)
    page_count += 1
    
    if json_data['hasMore'] == False:
        break
        
    res_jeju_data = res_jeju_data.append(pd_result)
```
